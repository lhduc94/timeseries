```python
df = pd.read_csv('../data/us-retail-sales.csv')
df['Month'] = pd.to_datetime(df['Month'])
df.head()
```

```python
df.Month>=pd.to_datetime('2022-01-01')
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Month', y='FoodAndBeverage', data=df[(pd.to_datetime('2008-01-01')>=df.Month) & (df.Month>=pd.to_datetime('2000-01-01'))])
plt.xlabel('Month',fontsize=15)
plt.ylabel('FoodAndBeverage',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

```
